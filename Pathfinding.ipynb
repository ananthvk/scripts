```python
class PathFinder:
    def __init__(self,loc_map):
        self.loc_map = loc_map
        self.grid = [[j for j in i if j.strip()!=''] for i in loc_map.split('\n') if i]
        self.nodes = [[Node(char=col) for col in row] for row in self.grid]
    
    def display(self):
        for y,row in enumerate(self.nodes):
            for x,node in enumerate(row):
                print(node,end=' ')
            print()
    def connect_edges(self):
        # Connects the individual nodes
        # None means there is no Node connected in that direction.
        # TODO: Probably there is way by which when you set that one is the right of 
        # the other, you can also set that the other one is left of the current node
        # , this can save computation time.
        # Look into it later.
        
        for i in range(len(self.nodes)):
            for j in range(len(self.nodes[0])):
                
                self.nodes[i][j].y = i
                self.nodes[i][j].x = j
                
                # Links to the node above the current node
                # If currently in the topmost row, don't link
                if i != 0:
                    self.nodes[i][j].up = self.nodes[i-1][j]

                # Links to the node below the current node
                # If currently in the bottom most row, don't link
                if i != len(self.nodes) - 1: 
                    self.nodes[i][j].down = self.nodes[i+1][j]
                
                # Links to the node right of the current node
                # If the current column is the last column, don't link
                if j != len(self.nodes[0]) - 1:
                    self.nodes[i][j].right = self.nodes[i][j+1]
                
                # Links to the node left of the current node
                # If the current column is the first column, don't link
                if j != 0:
                    self.nodes[i][j].left = self.nodes[i][j-1]
                    
    def does_path_exist(self,end, start = (0,0)):
        start_node = self.nodes[start[0]][start[1]]
        end_node = self.nodes[end[0]][end[1]]
        # Uses depth first serach to find if path exists or not
        visited = []
        backtracking_stack = [start_node]
        while backtracking_stack:
            current = backtracking_stack[-1]
            if current == end_node:
                # The path has been found
                return True
            choices = [current.up, current.down, current.left, current.right]
            choices = [i for i in choices if i is not None]
            cont = False
            for choice in choices:
                cont = False
                # Check if the choice has been visited
                if choice not in visited:
                    #print(f'Visiting {choice.__repr__()}')
                    visited += [choice]
                    #if choice == end_node:
                    #    # The path has been found
                    #    return True
                    if choice.char == '.':
                        # A floor, able to move.
                        #print(f'Moving to node {choice.__repr__()}')
                        backtracking_stack.append(choice)
                        cont = True
                        break
            if cont:
                continue
            # If the control reaches here, It means that all choices are visited
            # Or all the choices are walls.
            # We have to backtrack
            backtracking_stack.pop()
        
        # Path not found
        return False
                       
        
    def get_last_indice(self):
        return (len(self.nodes) - 1, len(self.nodes[0]) - 1)
```
